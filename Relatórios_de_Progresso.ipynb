```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados
dados = {
    "segunda": {"horas_trabalhadas": 6, "bugs_corrigidos": 3, "tarefas_concluidas": 3},
    "terça": {"horas_trabalhadas": 7, "bugs_corrigidos": 2, "tarefas_concluidas": 4},
    "quarta": {"horas_trabalhadas": 8, "bugs_corrigidos": 1, "tarefas_concluidas": 6},
    "quinta": {"horas_trabalhadas": 6, "bugs_corrigidos": 4, "tarefas_concluidas": 4},
    "sexta": {"horas_trabalhadas": 7, "bugs_corrigidos": 3, "tarefas_concluidas": 5},
    "sábado": {"horas_trabalhadas": 5, "bugs_corrigidos": 2, "tarefas_concluidas": 3},
    "domingo": {"horas_trabalhadas": 4, "bugs_corrigidos": 1, "tarefas_concluidas": 2}
}

# Convertendo os dados para DataFrame
df = pd.DataFrame.from_dict(dados, orient='index')

# Análise Exploratória

## Visualização dos dados
df.head()

## Descrição estatística dos dados
df.describe()

# Cálculos

## Total e média de horas trabalhadas
total_horas_trabalhadas = df['horas_trabalhadas'].sum()
media_horas_trabalhadas = df['horas_trabalhadas'].mean()

## Total e média de bugs corrigidos
total_bugs_corrigidos = df['bugs_corrigidos'].sum()
media_bugs_corrigidos = df['bugs_corrigidos'].mean()

## Total e média de tarefas concluídas
total_tarefas_concluidas = df['tarefas_concluidas'].sum()
media_tarefas_concluidas = df['tarefas_concluidas'].mean()

## Calculando produtividade diária
df['produtividade'] = df['tarefas_concluidas'] / df['horas_trabalhadas']

# Visualizações

## Gráfico de horas trabalhadas por dia
df['horas_trabalhadas'].plot(kind='bar', figsize=(10,6), color='orange')
plt.title('horas trabalhadas por dia')
plt.xlabel('dia')
plt.ylabel('horas trabalhadas')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

## Gráfico de bugs corrigidos por dia
df['bugs_corrigidos'].plot(kind='bar', figsize=(10,6), color='green')
plt.title('bugs corrigidos por dia')
plt.xlabel('dia')
plt.ylabel('bugs corrigidos')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

## Gráfico de tarefas concluídas por dia
df['tarefas_concluidas'].plot(kind='bar', figsize=(10,6), color='blue')
plt.title('tarefas concluídas por dia')
plt.xlabel('dia')
plt.ylabel('tarefas concluídas')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

## Gráfico de produtividade diária
df['produtividade'].plot(kind='bar', figsize=(10,6), color='skyblue')
plt.title('produtividade diária')
plt.xlabel('dia')
plt.ylabel('tarefas concluídas por hora')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

## Gráfico sintetizado de horas trabalhadas, bugs corrigidos e tarefas concluídas por dia
df.plot(kind='bar', figsize=(12,8))
plt.title('resumo diário')
plt.xlabel('dia')
plt.ylabel('quantidade')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.legend(['horas trabalhadas', 'bugs corrigidos', 'tarefas concluídas'])
plt.show()

# Relatório

print("# relatório de progresso semanal - Projeto de desenvolvimento de software\n")
print("## Período: segunda a domingo\n")
print(f"Total de horas trabalhadas: {total_horas_trabalhadas} horas")
print(f"Média diária de horas trabalhadas: {media_horas_trabalhadas:.2f} horas\n")
print(f"Total de bugs corrigidos: {total_bugs_corrigidos} bugs")
print(f"Média diária de bugs corrigidos: {media_bugs_corrigidos:.2f} bugs\n")
print(f"Total de tarefas concluídas: {total_tarefas_concluidas} tarefas")
print(f"Média diária de tarefas concluídas: {media_tarefas_concluidas:.2f} tarefas\n")

```
