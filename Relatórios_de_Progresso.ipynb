```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados
dados = {
    "segunda": {"horas_trabalhadas": 6, "bugs_corrigidos": 3, "tarefas_concluidas": 3},
    "tera": {"horas_trabalhadas": 7, "bugs_corrigidos": 2, "tarefas_concluidas": 4},
    "quarta": {"horas_trabalhadas": 8, "bugs_corrigidos": 1, "tarefas_concluidas": 6},
    "quinta": {"horas_trabalhadas": 6, "bugs_corrigidos": 4, "tarefas_concluidas": 4},
    "sexta": {"horas_trabalhadas": 7, "bugs_corrigidos": 3, "tarefas_concluidas": 5},
    "sbado": {"horas_trabalhadas": 5, "bugs_corrigidos": 2, "tarefas_concluidas": 3},
    "domingo": {"horas_trabalhadas": 4, "bugs_corrigidos": 1, "tarefas_concluidas": 2}
}

# Convertendo os dados para DataFrame
df = pd.DataFrame.from_dict(dados, orient='index')

# Anlise Exploratria

## Visualizao dos dados
df.head()

## Descrio estatstica dos dados
df.describe()

# Clculos

## Total e mdia de horas trabalhadas
total_horas_trabalhadas = df['horas_trabalhadas'].sum()
media_horas_trabalhadas = df['horas_trabalhadas'].mean()

## Total e mdia de bugs corrigidos
total_bugs_corrigidos = df['bugs_corrigidos'].sum()
media_bugs_corrigidos = df['bugs_corrigidos'].mean()

## Total e mdia de tarefas concludas
total_tarefas_concluidas = df['tarefas_concluidas'].sum()
media_tarefas_concluidas = df['tarefas_concluidas'].mean()

## Calculando produtividade diria
df['produtividade'] = df['tarefas_concluidas'] / df['horas_trabalhadas']

# Visualizaes

## Grfico de horas trabalhadas por dia
df['horas_trabalhadas'].plot(kind='bar', figsize=(10,6), color='orange')
plt.title('horas trabalhadas por dia')
plt.xlabel('dia')
plt.ylabel('horas trabalhadas')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

## Grfico de bugs corrigidos por dia
df['bugs_corrigidos'].plot(kind='bar', figsize=(10,6), color='green')
plt.title('bugs corrigidos por dia')
plt.xlabel('dia')
plt.ylabel('bugs corrigidos')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

## Grfico de tarefas concludas por dia
df['tarefas_concluidas'].plot(kind='bar', figsize=(10,6), color='blue')
plt.title('tarefas concludas por dia')
plt.xlabel('dia')
plt.ylabel('tarefas concludas')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

## Grfico de produtividade diria
df['produtividade'].plot(kind='bar', figsize=(10,6), color='skyblue')
plt.title('produtividade diria')
plt.xlabel('dia')
plt.ylabel('tarefas concludas por hora')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

## Grfico sintetizado de horas trabalhadas, bugs corrigidos e tarefas concludas por dia
df.plot(kind='bar', figsize=(12,8))
plt.title('resumo dirio')
plt.xlabel('dia')
plt.ylabel('quantidade')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.legend(['horas trabalhadas', 'bugs corrigidos', 'tarefas concludas'])
plt.show()

# Relatrio

print("# relatrio de progresso semanal - Projeto de desenvolvimento de software\n")
print("## Perodo: segunda a domingo\n")
print(f"Total de horas trabalhadas: {total_horas_trabalhadas} horas")
print(f"Mdia diria de horas trabalhadas: {media_horas_trabalhadas:.2f} horas\n")
print(f"Total de bugs corrigidos: {total_bugs_corrigidos} bugs")
print(f"Mdia diria de bugs corrigidos: {media_bugs_corrigidos:.2f} bugs\n")
print(f"Total de tarefas concludas: {total_tarefas_concluidas} tarefas")
print(f"Mdia diria de tarefas concludas: {media_tarefas_concluidas:.2f} tarefas\n")

```
